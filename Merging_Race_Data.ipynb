```python
import pandas as pd
```

```python
# File to Load
race_path = "Data_Racing/International_Data/races_2020.csv"
horse_path = "Data_Racing/International_Data/horses_2020.csv"

# Read the csv and store into Pandas DataFrame
race_df = pd.read_csv(race_path, encoding="utf-8")
race_df.head()

horse_df = pd.read_csv(horse_path, encoding="utf-8")
horse_df.head()
```

## Merging both datasets

```python
merge_data_df = race_df.merge(horse_df, how='left', on='rid')
merge_data_df.head()
```

## Cleaning Data
