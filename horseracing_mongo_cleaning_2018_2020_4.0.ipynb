```python
from pymongo import MongoClient
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# confirm that our new database was created
print(mongo.list_database_names())
```

```python
# assign the database to a variable name
db_2020 = mongo['horse_racing']
db_2019 = mongo['horse_racing_2019']
db_2018 = mongo['horse_racing_2018']
```

```python
# review the collections in our new database
print(db_2020.list_collection_names())
print(db_2019.list_collection_names())
print(db_2018.list_collection_names())
```

```python
# review a document in the customer_list collection
print(db_2019.horses_2019.find_one())
```

```python
# review a document in the customer_list collection
print(db_2018.races_2018.find_one())
```

```python
horses_2020 = db_2020['horses']
races_2020 = db_2020['races']
horses_2019 = db_2019['horses_2019']
races_2019 = db_2019['races_2019']
horses_2018 = db_2018['horses_2018']
races_2018 = db_2018['races_2018']
```

```python
horse_2020 = list(horses_2020.find())
race_2020 = list(races_2020.find())
horse_2019 = list(horses_2019.find())
race_2019 = list(races_2019.find())
horse_2018 = list(horses_2018.find())
race_2018 = list(races_2018.find())
```

```python
horse_2020_df = pd.DataFrame(horse_2020)
horse_2019_df = pd.DataFrame(horse_2019)
horse_2018_df = pd.DataFrame(horse_2018)
horse_2018_df.head()
```

```python
race_2020_df = pd.DataFrame(race_2020)
race_2019_df = pd.DataFrame(race_2019)
race_2018_df = pd.DataFrame(race_2019)
race_2020_df.head()
```

```python
race_features = ['rid', 'ages', 'metric', 'ncond']
race_2020_filtered = race_2020_df[race_features]
race_2019_filtered = race_2019_df[race_features]
race_2018_filtered = race_2018_df[race_features]
race_2018_filtered
```

```python
horse_features = ['rid', 'age', 'saddle', 'decimalPrice', 'isFav', 'RPR', 'TR', 'OR', 'father', 'weight', 'res_win']
horse_2020_filtered = horse_2020_df[horse_features]
horse_2019_filtered = horse_2019_df[horse_features]
horse_2018_filtered = horse_2018_df[horse_features]
horse_2018_filtered
```

# Merge

```python
# join 2020 data
data_2020 = race_2020_filtered.merge(horse_2020_filtered, how='inner', on='rid')
data_2020 = data_2020.drop(columns=['rid'])
data_2020
```

```python
# join 2019 data
data_2019 = race_2019_filtered.merge(horse_2019_filtered, how='inner', on='rid')
data_2019 = data_2019.drop(columns=['rid'])
data_2019
```

```python
# join 2018 data
data_2018 = race_2018_filtered.merge(horse_2018_filtered, how='inner', on='rid')
data_2018 = data_2018.drop(columns=['rid'])
data_2018.head()
```

```python
# filter fathers 2020
father_counts_2020 = data_2020['father'].value_counts()
popular_sires_2020 = father_counts_2020.index[father_counts_2020 >= 100]
print(f'number of popular_sires: {len(popular_sires_2020)}')
data_2020 = data_2020[data_2020['father'].isin(popular_sires_2020)]
data_2020
```

```python
# filter fathers 2019
father_counts_2019 = data_2019['father'].value_counts()
popular_sires_2019 = father_counts_2019.index[father_counts_2019 >= 100]
print(f'number of popular_sires: {len(popular_sires_2019)}')
data_2019 = data_2019[data_2019['father'].isin(popular_sires_2019)]
data_2019
```

```python
# drop na
cleaned_data_2020 = data_2020.dropna()
cleaned_data_2019 = data_2019.dropna()
cleaned_data_2019
```

```python
# merge 2020 and 2019 data 
cleaned_2020_2019_df = pd.concat([cleaned_data_2019, cleaned_data_2020])
cleaned_2020_2019_df
```
