```python
# File to Load
race_path = "Data_Racing/International_Data/races_2020.csv"
horse_path = "Data_Racing/International_Data/horses_2020.csv"

# Read the csv and store into Pandas DataFrame
race_df = pd.read_csv(race_path, encoding="utf-8")
horse_df = pd.read_csv(horse_path, encoding="utf-8")
```

```python
horse_features = ['rid', 'age', 'saddle', 'decimalPrice', 'isFav', 'RPR', 'TR', 'OR', 'father', 'weight', 'res_win']
horse_df_filtered = horse_df[horse_features]
```

```python
horse_df_filtered
```

# Preprocess

```python
# join data
data_df = race_df_filtered.merge(horse_df_filtered, how='inner', on='rid')
data_df = data_df.drop(columns=['rid'])
data_df
```

```python
# filter fathers
father_counts = data_df['father'].value_counts()
popular_sires = father_counts.index[father_counts >= 100]
print(f'number of popular_sires: {len(popular_sires)}')
data_df = data_df[data_df['father'].isin(popular_sires)]
data_df
```

```python
# drop na
cleaned_data_df = data_df.dropna()
cleaned_data_df
```
